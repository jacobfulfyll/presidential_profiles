```python
import pandas as pd

df = pd.read_csv('../data/speeches_by_paragraph.csv')
```

```python
full_doc_df = df[df['full_document'] == 1]
paragraphs_df = df[df['full_document'] == 0]
full_doc_df.reset_index()
paragraphs_df.reset_index()
paragraphs_df = paragraphs_df.drop(columns='full_document')
paragraphs_df
```

```python
new_df = pd.DataFrame(columns=['date', 'speech_title', 'president', 'text', 'paragraph'])
counter = 0
d_counter = 0
for idx, d in full_doc_df['text'].iteritems():
    print(d_counter, ' / ', len(full_doc_df))
    d = d.split('\n')
    date = full_doc_df.loc[idx]['date']
    title = full_doc_df.loc[idx]['speech_title']
    president = full_doc_df.loc[idx]['president']
    p_counter = 0
    for p in d:
        if p == '':
            pass
        else:
            new_df.loc[counter] = [date, title, president, p, p_counter]
            counter += 1
            p_counter += 1
    d_counter += 1
new_df
```

```python
final_df = pd.concat([paragraphs_df, new_df])
final_df = final_df.drop(columns='Unnamed: 0')
final_df
```

```python
final_df.to_csv('../data/speeches.csv')
```
